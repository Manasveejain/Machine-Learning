### N arguments

```python
#multiply 2 numbers

def mul(a, b , z):
    return a*b *z
```

```python
mul(2,3,4)
```

```python
#multiply 2 numbers

def mul(*numbers):
    ans = 1
    for num in numbers:
        ans = ans * num
    return ans
```

```python
mul(2,3,4,5)
```

```python
mul(2,11)
```

```python
mul(7,8,9,2,3,4)
```

### Lambda Function

```python
# anonymous function

#Lambda parameter : expression

lambda a,b,c : a + b + c
```

```python
square = lambda x : x**2
square(5)
```

### Zip functions

```python
first_names = ["john", "Mary", "Bob"]
last_names = ["smith", "Jane" ,"smith"]

full_names = list(zip(first_names, last_names))

full_names
```

### Map

```python
def square(x):
    return x ** 2

numbers =[2,3,4,5]

sq = list(map(square, numbers))
sq
```
