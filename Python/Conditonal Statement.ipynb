```python
temperature = 18

if temperature > 20 :
    print("Its a warm Day")
else:
    print("Its a cold Day")
```

### if - elif statement

```python
temperature = 35

if temperature > 30 :
    print("Its a Hot Day")
elif temperature >20:
    print("Its a warm Day")
else:
    print("Its a cold Day")
```

### Nested Condition

```python
temperature = 13

weather ="sunny"

if temperature >15:
    if weather == "sunny":
        print("Its a sunny and warm Day")
    else:
        print("Its a warm day but not sunny")
else:
    print("Its a not a warm Day")
```

## Loops

```python
# List of fruits

fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
```

```python
numbers =[1,2,3,4,5]

for num in numbers:
    print(num)
```

```python
for num in range(1, 101):
    print(num)
```

```python
course = "Python Programming"

for letter in course:
    print(letter)
```

## Nested Loop

```python
sizes = ["small", "medium" ,"large"]
clothes = ["shirt", "Pant"]

for size in sizes:
    for cloth in clothes:
        print(size, cloth)
```

### while loop

```python
number =10

while number > 0:
    print(number)
    number -= 1
```

```python
number = 10
while number > 0:
    if number == 4:
        break
    print(number)
    number -= 1
```
