### Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")
```

```python
#Load the dataset
df = pd.read_csv('sample_data.csv')
```

```python
df.head()
```

```python
# set style of plot
sns.set(style = 'whitegrid')
```

### Histogram

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Potential'], bins = 20, kde = True , color = 'purple')
plt.title('Distribution of potential')
plt.show()
```

### Scatter plot with Hue

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x = "Potential" , y ='Age', data = df , hue = 'Nationality', palette = 'deep')
plt.title('Potential vs Age')
plt.show()
```

### Box plot

```python
plt.figure(figsize=(6,4))
sns.boxplot(x = "Potential" , y ='Age', data = df, palette = 'pastel')
plt.title('Potential vs Age')
plt.show()
```

```python
df1 = pd.read_csv('sample_data1.csv')
df1.head()
```

### Bar Plot

```python
plt.figure(figsize=(10,4))
sns.barplot(x = "Year" , y ='Italy', data = df1,estimator = 'mean', palette = 'muted')
plt.title('gas prices in Italy')
plt.xticks(rotation= 45)
plt.show()
```

### Pair plot


```python
plt.figure(figsize=(6,4))
sns.pairplot(df[['Potential' ,'Overall' , 'Age']] , diag_kind = 'kde')
plt.show()
```

### HeatMap

```python
plt.figure(figsize=(6,4))
corr_matrix = df[['Potential' ,'Overall' , 'Age']].corr()
sns.heatmap(corr_matrix , annot = True, cmap ='coolwarm', fmt ='.2f')
plt.title('correlation heatmap')
plt.show()
```
