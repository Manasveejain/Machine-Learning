### Load necessary libraries


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x=[0,1,2,3,4]
y= [0,2,4,6,8]

# Resize your graph

plt.figure(figsize=(5,3) , dpi=300)
#line 1

#Keyword Argument Notation
#plt.plot(x,y, label ='2x', color='red', linewidth = 2, linestyle ='--',
#          marker='.', markersize=10, markeredgecolor = 'blue')

#shorthand notation
#fmt = '[color][marker][line]'
# plt.plot(x,y, 'r.--', label ='2x')
plt.plot(x,y, 'b^--', label ='2x')


## Line 2

# select interval as we want to plot point at
x2 = np.arange(0,4.5,0.5)

#plot part of the graph as line
plt.plot(x2[:6] , x2[:6]**2, 'r', label ='X^2')

#plot remaining graph
plt.plot(x2[5:] , x2[5:]**2, 'r--' )

# Add a title (specify font parameter with fontdict)
plt.title('Our First Graph', fontdict ={'fontname':'Comic Sans MS', 
                'fontsize':20})

#X and Y labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# X and y tickmarks (scale of graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#add a legend
plt.legend()

#save figure (dpi 300)
plt.savefig('mygraph.png', dpi =300)

#show plot
plt.show()
```

### Bar chart

```python
labels =['A','B','C']
values =[1,4,2]

bars = plt.bar(labels,values)

patterns =['/','o','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.figure(figsize =(6,4))

plt.show()
```

### Histogram

```python
# create a dataset for histogram 
np.random.seed(0)  #reproductivity 
data = np.random.normal(loc=0, scale = 1, size = 1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=30, color='purple', edgecolor = 'black')
plt.title('Random data distribution')
plt.xlabel('value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
# create a dataset for scatter plot 
np.random.seed(1)  #reproductivity 

x= np.random.rand(50)*100  # random x values
y= np.random.rand(50)*100   # random y values

plt.figure(figsize=(6,4))
plt.scatter(x, y ,color='red')
plt.title('Random scatter plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()
```

### Box plot

```python
# create a dataset for Box plot 
np.random.seed(2)  #reproductivity 

group1 = np.random.normal(loc = 0 , scale =1 , size = 100)  # Data for group 1
group2 = np.random.normal(loc = 0 , scale =1.5 , size = 100)   # Data for group 2

plt.figure(figsize=(6,4))
plt.boxplot([group1 , group2], labels = ['group1' , 'group2'])
plt.title('Box plot of 2 groups')
plt.xlabel('Group ')
plt.ylabel('value')
plt.grid(True)
plt.show()
```

### Heat Map

```python
# create a correlation matrix for demonstration
np.random.seed(3)  #reproductivity 

data_matrix =  np.random.rand(10, 10)

plt.figure(figsize=(6,4))
plt.imshow(data_matrix, cmap ='coolwarm', interpolation ='none')
plt.colorbar(label = 'value')
plt.title('Heatmap of random Data matrix')
plt.show()
```

### creating subplots

```python
# Generate some simple data using arrays
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

# create a figure
plt.figure(figsize =(6,2))

# first subplot : sine wave 
plt.subplot(1,2,1)
plt.plot(x, y1, 'b--')
plt.title('sine wave')

# sec subplot :
plt.subplot(1,2,2)
plt.plot(x,y2, 'r-')
plt.title('cosine wave')


plt.tight_layout()
plt.show()
```
