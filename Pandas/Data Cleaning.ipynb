```python
import pandas as pd
import numpy as np 
```

```python
data ={
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"], 
    "Age": [25, 30, 35, np.nan, 29, 25], 
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"], 
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000]
}
```

```python
df =pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna()  #drop null values
```

```python
df.dropna(how ="any") #any row that had null value
```

```python
df.dropna(how = 'all')  #if all the values in a row is null
```

```python
df.fillna(0) #NOOB Method
```

```python
# Fill missing values in stock with the mean of the column
df['Age'].fillna(df["Age"].mean())
```

```python
# Fill missing values in stock with the median of the column
df['Salary'].fillna(df["Salary"].median())
```

```python
df['Age'].fillna(method = "ffill")
```

```python
df["Name"]=df["Name"].replace("Charlie","Rose")
```

```python
df
```

### Duplicates:

```python
df_dup = df[df.duplicated( keep ="last")]
```

```python
df_dup 
```

```python
# Remove duplicate rowa
df = df.drop_duplicates()
```

```python
df
```

### Invalid Values

```python
name  = "alice _fernandes"
df[["first_name","last_name"]] = df["Name"].str.splt("_")
```
