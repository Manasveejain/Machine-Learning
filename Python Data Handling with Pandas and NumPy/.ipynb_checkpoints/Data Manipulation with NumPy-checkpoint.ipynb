## Task 1: NumPy Array Operations
○ Objective: Perform operations on NumPy arrays.
○ Instructions:
1. Create a NumPy array with 12 random integers between 1 and 50.
2. Reshape the array into a 3x4 matrix.
3. Find and print the maximum, minimum, and mean values of the array

```python
import numpy as np
```

```python
arr = np.random.randint(1, 51, size=12)
print(arr)
```

```python
new_arr = arr.reshape(3,4)
print(new_arr)
```

```python
max_value = arr.max()
print(max_value)
```

```python
min_value = arr.min()
print(min_value)
```

```python
mean_value = arr.mean()
print(mean_value)
```

## Task 2: Array Indexing and Slicing
○ Objective: Practice indexing and slicing with NumPy arrays.
○ Instructions:
1. Create a NumPy array of shape (4, 3) with values from 1 to 12.
2. Extract and print the second row of the array.
3. Extract and print a 2x2 submatrix from the top-left corner of the array.

```python
arr2 = np.arange(1,13).reshape(4,3)
print(arr2)
```

```python
print(arr2[1 ,:])
```

```python
print(arr2[:2, :2])
```
