```python
import pandas as pd
import numpy as np
```

## Task 1: Creating and Exploring a DataFrame
○ Objective: Create a Pandas DataFrame from a dictionary and explore its contents.
○ Instructions:
1. Create a DataFrame from this dictionary.
2. Display the first few rows of the DataFrame.
3. Calculate and print the average age from the DataFrame

```python
data = {
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Age': [25, 30, 35, 40, 45],
'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
mean = df['Age'].mean()
print(mean)
```

## Task 2: Data Filtering and Aggregation
○ Objective: Filter and aggregate data in a DataFrame.
○ Instructions:
1. Filter the DataFrame to include only rows where the age is greater than 30.
2. Add a new column to the DataFrame called AgeGroup, where the value is 'Senior' if age is greater than or
equal to 40, otherwise 'Adult'.
3. Group the data by AgeGroup and calculate the average age for each group.

```python
df_filtered = df[df['Age'] > 30].copy()
print(df_filtered)
```

```python
def ageGroup(age):
    if age >= 40:
        return 'Senior'
    return 'Adult'
    
```

```python
df["AgeGroup"] = df["Age"].apply(ageGroup)
```

```python
df['AgeGroup']
```

```python
df_filtered['Age Group'] = np.where(df_filtered['Age'] >= 40, 'Senior', 'Adult')
print(df_filtered)
```

```python
average_ages = df_filtered.groupby('Age Group')['Age'].mean()
print(average_ages)
```
